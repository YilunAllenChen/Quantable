```python
import pandas as pd
import numpy as np

prices = pd.read_pickle("./prices.pkl")


data = prices.pct_change()


correlations = data.corr()
```

```python

# Find cells where the value is less than 0.8
mask = correlations > 0.9

# Get the row and column labels of these cells
rows, cols = np.where(mask)

# Print the row and column labels
for row, col in zip(rows, cols):
    if row != col:
        print("Row:", correlations.index[row], "Column:", correlations.columns[col])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='white')

mask = np.triu(np.ones_like(correlations, dtype=bool))

f, ax = plt.subplots(figsize=(20, 20))

cmap = sns.diverging_palette(230, 20, as_cmap=True)

sns.heatmap(correlations, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```
